```python
import numpy as np
from scipy.stats import rv_discrete

# 支持点
x = np.array([1, 2, 3, 4])

# 概率质量函数
pmf = (5 - x) / 10

# 构造离散随机变量
X = rv_discrete(values=(x, pmf))

# 计算期望值
expected_value = X.mean()

print(round(expected_value, 1))
```
